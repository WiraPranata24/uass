```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE
import joblib
```

```python
file_path = '/content/podcasts.csv'
data = pd.read_csv(file_path)
```

```python
data_info = data.info()
data_head = data.head()
print(data_info)
print(data_head)
```

```python
#preprocessing
data['languages'] = data['languages'].str.strip("[]").str.replace("'", "").str.split(", ")
data['languages'] = data['languages'].apply(lambda x: x[0] if x else "unknown")
data['publisher'] = data['publisher'].str.lower()
data['media_type'] = data['media_type'].str.lower()
```

```python
X = data[['total_episodes', 'languages', 'media_type', 'publisher']]
y = data['explicit']
```

```python
encoder = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'), ['languages', 'media_type', 'publisher'])
], remainder='passthrough')
X_encoded = encoder.fit_transform(X)
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)
```

```python
#split dataset
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
#train klasifikasi random forest
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
#prediksi
y_pred = rf_classifier.predict(X_test)
```

```python
#metriks evaluasi
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, rf_classifier.predict_proba(X_test)[:, 1]))
```

```python
joblib.dump(rf_classifier, "random_forest_model.pkl")
joblib.dump(encoder, "encoder.pkl")
```
